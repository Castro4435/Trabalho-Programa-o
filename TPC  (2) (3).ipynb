```python
#import pandas in jupyter notebook:
import pandas as pd



df_fam= pd.read_excel("PORDATA_Por-tipo-de-composição (1).xlsx")
df_fam.head()
```

```python
#Fizemos isto porque no Excel apareciam imensas colunas sem dados, todos com NaN
```

```python
df_ok=df_fam[["Anos", "Total","1 indivíduo","Casal sem filhos","Casal com filhos","Familia monoparental","Outros"]]
df_ok.head()
```

```python
df_ok.tail()
```

```python
df_mono= pd.read_excel("PORDATA_Monoparentais-por-sexo (1).xlsx")
df_mono.head()
```

```python
df_monol=df_mono[["Anos","Total","Masculino","Feminino"]]
df_monol.head()
```

```python
#Adicionamos este passo de modo a que sejam incluidos somente as colunas que possuem dados
```

```python
df_monol2 = df_monol[df_monol.index<30]
df_monol2
```

```python
df_monol2.info()
```

```python
#Aqui alteramos tabelas de objetos para valores numericos inteiros

df_monol2['Anos'] = df_monol2['Anos'].astype(int)
df_monol2['Total'] = df_monol2['Total'].astype(int)
df_monol2['Masculino'] = df_monol2['Masculino'].astype(int)
df_monol2['Feminino'] = df_monol2['Feminino'].astype(int)
df_monol2.dtypes
```

```python
#Adicionamos este passo de modo a criar a media de familias onde apenas esta presente o pai

result=df_monol2.describe()
result
```

```python
average_masc=result.loc["mean","Masculino"]
average_masc
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
# Set figure default figure size
plt.rcParams["figure.figsize"] = (12, 6)

plt.plot('Anos', "Total",  data=df_ok, linestyle='-', marker='')
plt.plot('Anos', "1 indivíduo",  data=df_ok, linestyle='-', marker='')
plt.plot('Anos', "Casal sem filhos",  data=df_ok, linestyle='-', marker='')
plt.plot('Anos', "Casal com filhos",  data=df_ok, linestyle='-', marker='')
plt.plot('Anos', "Familia monoparental",  data=df_ok, linestyle='-', marker='')
plt.plot('Anos', "Outros",  data=df_ok, linestyle='-', marker='')

# show legend
plt.legend(bbox_to_anchor =(1., 1.))

plt.show()
```
